```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return np.exp(np.cos(x))
```

```python
x = np.arange(0.1, 1.1, 0.1)
y = np.array(func(x))

print(f'x = {x}')
print('y = [ ', end='')
for el in y:
    print(f'{el:.4f} ', end='')
print(']')
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

mean_x2 = np.mean(x**2)
mean_xy = np.mean(x*y)

print(f'mean_x  = {mean_x:.4f}\tmean_y  = {mean_y:.4f}')
print(f'mean_x2 = {mean_x2:.4f}\tmean_xy = {mean_xy:.4f}')
```

```python
a1 = (mean_xy - mean_x * mean_y) / (mean_x2 - mean_x**2)
a0 = mean_y - a1 * mean_x

print(f'a1 = {a1:.4f}\ta0 = {a0:.4f}')
```

```python
plt.plot(x, a0 + a1 * x, label='Апроксимована пряма')
plt.plot(x, y, label='Функція')
plt.title('Метод найменших квадратів')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.linspace(0.1, 1, 10)
y = np.array(func(x))

print(f'x = {x}')
print('y = [ ', end='')
for el in y:
    print(f'{el:.4f} ', end='')
print(']')
```

```python
a = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(a, y, rcond=None)[0]

print(f'a0 = {c:.4f}\ta1 = {m:.4f}')
print(f'y = {c:.4f}{m:+.4f}*x')
```

```python
plt.plot(x, m * x + c, label='Апроксимована пряма')
plt.plot(x, y, label='Функція')    
plt.title('Метод найменших квадратів')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
